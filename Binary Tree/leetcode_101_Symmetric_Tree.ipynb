```python
from typing import List, Optional
from collections import deque
```

[leetcode 101. Symmetric Tree](https://leetcode.com/problems/symmetric-tree/description/)

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution:
    def solve(self, r1: Optional[TreeNode], r2: Optional[TreeNode]) -> bool:
        if r1 is None and r2 is None:
            return True

        if r1 is None or r2 is None:
            return False

        if r1.val != r2.val:
            return False

        return self.solve(r1.left, r2.right) and self.solve(r1.right, r2.left)


    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
       return self.solve(root.left, root.right)
```

```python
class Solution:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        q = deque([root])

        while q:
            size = len(q)
            nodes = []

            for _ in range(size):
                node = q.popleft()

                if node.left:
                    q.append(node.left)
                    nodes.append(node.left.val)
                else:
                    nodes.append(None)

                if node.right:
                    q.append(node.right)
                    nodes.append(node.right.val)
                else:
                    nodes.append(None)

            for j in range(len(nodes)//2):
                if nodes[j] != nodes[len(nodes)-j-1]:
                    return False

        return True
```
