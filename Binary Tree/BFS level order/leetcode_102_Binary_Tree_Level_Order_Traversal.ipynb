[leetcode 102. Binary Tree Level Order Traversal](https://leetcode.com/problems/binary-tree-level-order-traversal/description/)

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if root is None:
            return []

        q = deque([root])
        result = []

        while q:
            size = len(q)
            nodes = [None] * size

            for i in range(size):
                node = q.popleft()
                nodes[i] = node.val

                if node.left:
                    q.append(node.left)

                if node.right:
                    q.append(node.right)

            result.append(nodes)

        return result
```
