```python
from typing import Optional
from collections import deque
```

[leetcode 112. Path Sum](https://leetcode.com/problems/path-sum/description/)

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution:
    def solve(self, root: Optional[TreeNode], path_sum: int, target: int) -> bool:
        if root is None:
            return False

        path_sum += root.val

        if path_sum == target and root.left is None and root.right is None:
            return True

        return self.solve(root.left, path_sum, target) or self.solve(root.right, path_sum, target)


    def hasPathSum(self, root: Optional[TreeNode], target: int) -> bool:
        return self.solve(root, 0, target)
```

```python
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], target: int) -> bool:
        if root is None:
            return False

        q = deque([(root, root.val)])

        while q:

            node, path_sum = q.popleft()

            if path_sum == target and node.left is None and node.right is None:
                return True

            if node.left:
                q.append((node.left, path_sum + node.left.val))

            if node.right:
                q.append((node.right, path_sum + node.right.val))

        return False
```
