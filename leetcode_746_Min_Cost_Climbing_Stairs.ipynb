```python
class Solution:
    def min_cost_from_step(self, idx: int, costs: List[int]) -> int:
        # Base case: reached or crossed the top
        if idx >= len(costs):
            return 0

        # Cost if we take 1 step or 2 steps
        cost_1_step = self.min_cost_from_step(idx + 1, costs)
        cost_2_steps = self.min_cost_from_step(idx + 2, costs)

        return costs[idx] + min(cost_1_step, cost_2_steps)

    def minCostClimbingStairs(self, costs: List[int]) -> int:
        start_from_0 = self.min_cost_from_step(0, costs)
        start_from_1 = self.min_cost_from_step(1, costs)

        return min(start_from_0, start_from_1)
```

```python
class Solution:
    def min_cost_from_step(self, idx: int, costs: List[int], dp: List[int]) -> int:
        # Base case: reached or crossed the top
        if idx >= len(costs):
            return 0

        # check if cost already computed for an idx
        if dp[idx] != -1:
            return dp[idx]

        # Cost if we take 1 step or 2 steps
        cost_1_step = self.min_cost_from_step(idx + 1, costs, dp)
        cost_2_steps = self.min_cost_from_step(idx + 2, costs, dp)

        # store the min cost of an idx
        dp[idx] = costs[idx] + min(cost_1_step, cost_2_steps)
        return dp[idx]

    def minCostClimbingStairs(self, costs: List[int]) -> int:
        n = len(costs)
        dp = [-1] * n
        self.min_cost_from_step(0, costs, dp)

        return min(dp[0], dp[1])
```

```python
class Solution:

    def minCostClimbingStairs(self, costs: List[int]) -> int:
        n = len(costs)
        dp = [-1] * (n + 2)

        # Base cases: n is Top reached and n+1 is Top crossed
        dp[n], dp[n+1] = 0, 0

        for i in range(n-1, -1, -1):
            dp[i] = costs[i] + min(dp[i+1], dp[i+2])

        return min(dp[0], dp[1])
```

```python
class Solution:
    def minCostClimbingStairs(self, costs: List[int]) -> int:
        n = len(costs)

        # Base cases: prev1 is Top reached and prev2 is Top crossed
        prev2, prev1 = 0, 0

        for i in range(n-1, -1, -1):
            prev2, prev1 = prev1, costs[i] + min(prev1, prev2)

        return min(prev1, prev2)
```
