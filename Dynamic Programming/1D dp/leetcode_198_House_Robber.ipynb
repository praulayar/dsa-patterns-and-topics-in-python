```python
class Solution:
    def max_rob_at_idx(self, idx: int, nums: List[int]) -> int:
        if idx >= len(nums):
            return 0
        take = nums[idx] + self.solve(idx+2, nums)
        skip = self.solve(idx+1, nums)
        return max(take, skip)

    def rob(self, nums: List[int]) -> int:
        return self.max_rob_at_idx(0, nums)
```

```python
class Solution:
    def max_rob_at_idx(self, idx: int, nums: List[int], dp: List[int]) -> int:
        if idx >= len(nums):
            return 0

        if dp[idx] != -1:
            return dp[idx]

        take = nums[idx] + self.max_rob_at_idx(idx+2, nums, dp)
        skip = self.max_rob_at_idx(idx+1, nums, dp)

        dp[idx] =  max(take, skip)
        return dp[idx]

    def rob(self, nums: List[int]) -> int:
        n = len(nums)
        dp = [-1] * n
        return self.max_rob_at_idx(0, nums, dp)
```

```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        n = len(nums)

        if n == 1:
            return nums[0]

        dp = [-1] * (n + 2)
        dp[n + 1], dp[n] = 0, 0

        for i in range(n - 1, -1, -1):
            dp[i] = max(nums[i] + dp[i + 2], dp[i + 1])

        return dp[0]
```

```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        n = len(nums)

        if n == 1:
            return nums[0]

        prev2, prev1 = 0, 0

        for i in range(n - 1, -1, -1):
            prev2, prev1 = prev1, max(nums[i] + prev2, prev1)

        return prev1
```
